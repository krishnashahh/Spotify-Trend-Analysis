```python
# Load dataset
df = pd.read_csv('spotifytoptracks.csv')  # Adjust the path accordingly
```

```python
# Check the first few rows
df.head()

# Check for missing values
df.isnull().sum()

# Fill or drop missing values if necessary
df.dropna(inplace=True)  # or df.fillna(method='ffill', inplace=True)
```

```python
print(df.columns)
```

```python
# Choose relevant numerical columns
features = ['energy', 'danceability', 'loudness', 'acousticness',
            'speechiness', 'instrumentalness', 'liveness', 'valence', 'tempo']

# Compute correlation
corr = df[features].corr()

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Song Features')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='genre', order=df['genre'].value_counts().index, palette='magma')
plt.title('Number of Songs per Genre')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='tempo', y='loudness', data=df, hue='genre', palette='tab10')
plt.title('Tempo vs Loudness by Genre')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Loudness (dB)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='energy', y='danceability', data=df, hue='genre', palette='Set2')
plt.title('Energy vs Danceability by Genre')
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
